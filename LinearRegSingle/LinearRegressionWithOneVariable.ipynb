```python
import matplotlib.pyplot as plt
from sklearn import linear_model
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('canada_per_capita_income.csv')
```

```python
data.head()
```

```python
plt.scatter(data.year,data['per capita income (US$)'],color='red')
plt.xlabel('year')
plt.ylabel('price')
```

```python
data[['year']]
```

```python
reg=linear_model.LinearRegression()
reg.fit(data[['year']],data['per capita income (US$)'])
```

```python
reg.predict([[2025]])
```

```python
reg.coef_
reg.intercept_
```

```python
 x = np.linspace(-10, 10, 100)
y = reg.coef_ * x + reg.intercept_
plt.plot(x, y, label=f'y = {reg.coef_}x + {reg.in}')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
ls=[]
for year in data.year:
    ls.append(reg.predict([[year]]))
ls
```

```python
%matplotlib inline
plt.scatter(data.year,data['per capita income (US$)'])
plt.plot(data.year,ls,color='red')
```

```python
data['new prices']=np.array(ls).ravel()
data.head()
```
