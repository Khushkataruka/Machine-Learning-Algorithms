```python
import pandas as pd
import numpy as np
df=pd.read_csv('carprices.csv')
df
```

```python
x=np.array(df[['Mileage','Age(yrs)']])
```

```python
y=np.array(df['Sell Price($)'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
reg.predict([[70000,6]])
```

```python
reg.score(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
reg.coef_
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['Mileage'], df['Sell Price($)'],color='blue')
plt.scatter(df['Age(yrs)'], df['Sell Price($)'], color='green')
predicted_salaries = reg.predict(df[['Mileage', 'Age(yrs)']])
plt.plot(x, predicted_salaries, label='Regression Line', color='black')
```
