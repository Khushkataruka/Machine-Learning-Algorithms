```python
import math

def gradientDescent(x, y):
    n = len(x)
    w = 0
    b = 0
    learning_rate = 1e-5
    iterations = 10000

    for i in range(iterations):
        # Prediction
        y_predicted = w * x + b

        # Compute cost (Mean Squared Error)
        cost = (1/2*n) * sum((y[i] - y_predicted[i])**2 for i in range(n))

        # Gradients
        wd = -(2/n) * sum(x[i] * (y[i] - y_predicted[i]) for i in range(n))
        bd = -(2/n) * sum(y[i] - y_predicted[i] for i in range(n))

        # Update weights and bias
        w -= learning_rate * wd
        b -= learning_rate * bd

        # Print the cost and weights for each iteration
        print(f"w: {w}, b: {b}, iteration: {i}, cost: {cost}")

        # Check for convergence
        if i > 0 and math.isclose(w, prev_w, rel_tol=1e-9) and math.isclose(b, prev_b, rel_tol=1e-9):
            return [w,b]
            break
        elif i==iterations-1:
              return [w,b]
        # Save current values to compare in the next iteration
        prev_w = w
        prev_b = b
```
